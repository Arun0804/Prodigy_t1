```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



```

```python
df=pd.read_csv('national-population.csv')
```

```python
df.head()
```

```python
df = pd.DataFrame(df)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['period'], df['population_change'], color='skyblue')
plt.xlabel('Period')
plt.ylabel('Population Change')
plt.title('Population Change Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['population'], bins=5, color='lightgreen', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution')
plt.show()
```
